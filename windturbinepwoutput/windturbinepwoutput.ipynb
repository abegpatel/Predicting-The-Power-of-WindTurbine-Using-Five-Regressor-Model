```python
#filling numeric no
s = (train.dtypes == 'float64')


object_cols = list(s[s].index)
print(object_cols)
trainc_pre = train.copy()
for col in object_cols:
    trainc_pre[col] = trainc_pre[col].fillna(train[col].mean())
trainc_pre.isna().sum()
```

```python
print(trainc_pre['turbine_status'].mode())
print(trainc_pre['cloud_level'].mode())
```

```python
#filling categorical var
trainc_pre['turbine_status']=trainc_pre.turbine_status.apply(lambda x: str(x).replace('nan','BB'))
trainc_pre['cloud_level']=trainc_pre['cloud_level'].apply(lambda x: str(x).replace('nan','LOW'))
```

```python
trainc_pre.isna().sum()
```

```python
#encodeing all categorical var
s = (train.dtypes == 'object')


object_cols = list(s[s].index)

print(object_cols.remove('tracking_id'))
print(object_cols.remove('datetime'))


print("Categorical variables:")
print(object_cols)

from sklearn.preprocessing import LabelEncoder

# Make copy to avoid changing original data 
traindata = trainc_pre.copy()

# Apply label encoder to each column with categorical data
label_encoder = LabelEncoder()
for col in object_cols:
    traindata[col] = label_encoder.fit_transform(trainc_pre[col])
```

```python
traindata
```

```python
#preprocess test set
test.isna().sum()
```

```python
train.info()
```

```python
#filling numeric no
s = (test.dtypes == 'float64')
object_cols = list(s[s].index)
print(object_cols)
testc_pre = test.copy()
for col in object_cols:
    testc_pre[col] = testc_pre[col].fillna(test[col].mean())
testc_pre.isna().sum()
```

```python
#filling categorical var
testc_pre['turbine_status']=testc_pre.turbine_status.apply(lambda x: str(x).replace('nan','BB'))
testc_pre['cloud_level']=testc_pre['cloud_level'].apply(lambda x: str(x).replace('nan','LOW'))
print(testc_pre['turbine_status'].mode())
print(testc_pre['cloud_level'].mode())
```

```python
testc_pre.isna().sum()
```

```python
#encodeing all categorical var
s = (test.dtypes == 'object')


object_cols1 = list(s[s].index)

print(object_cols1.remove('tracking_id'))
print(object_cols1.remove('datetime'))


print("Categorical variables:")
print(object_cols1)

from sklearn.preprocessing import LabelEncoder

# Make copy to avoid changing original data 
testdata = testc_pre.copy()

# Apply label encoder to each column with categorical data
label_encoder = LabelEncoder()
for col in object_cols1:
    testdata[col] = label_encoder.fit_transform(testc_pre[col])
    
```

```python
testdata
```

```python
#spliting data into x and y
X=traindata.iloc[:,2:-1]
y=traindata.iloc[:,-1]
#y1=label_data.iloc[:,-1].astype(str)
```

```python
X_train=traindata.iloc[:,2:-1]
y_train=traindata.iloc[:,-1]
X_test=testdata.iloc[:,2:]
```

```python
# Fitting Multiple Linear Regression to the Training set
from sklearn.linear_model import LinearRegression
regressor1 = LinearRegression()
regressor1.fit(X_train, y_train)
# Predicting the Test set results
y_pred1 = regressor1.predict(X_test)
ev1=regressor1.predict(X_train)
```

```python
# Fitting SVR to the dataset
from sklearn.svm import SVR
regressor3 = SVR(kernel = 'rbf')
regressor3.fit(X_train, y_train)
# Predicting the Test set results
y_pred3 = regressor3.predict(X_test)
ev3=regressor3.predict(X_train)
```

```python
# Fitting Decision Tree Regression to the dataset
from sklearn.tree import DecisionTreeRegressor
regressor4 = DecisionTreeRegressor(random_state = 0)
regressor4.fit(X_train, y_train)
# Predicting the Test set results
y_pred4 = regressor4.predict(X_test)
ev4=regressor4.predict(X_train)
```

```python
# Fitting Random Forest Regression to the dataset
from sklearn.ensemble import RandomForestRegressor
regressor5 = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor5.fit(X_train, y_train)
# Predicting the Test set results
y_pred5 = regressor5.predict(X_test)
ev5=regressor5.predict(X_train)
```
